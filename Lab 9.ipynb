```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("heart.csv")
```

```python
# -----------------------------
# a) Mean age
# -----------------------------
mean_age = df['age'].mean()
print("a) Mean Age:", round(mean_age, 2))
```

```python
# -----------------------------
# b) Heart disease counts
# -----------------------------
heart_counts = df['target'].value_counts()
print("\nb) Heart Disease Counts:")
print("With Heart Disease:", heart_counts.get(1, 0))
print("Without Heart Disease:", heart_counts.get(0, 0))
```

```python
# -----------------------------
# c) Max and Min Cholesterol
# -----------------------------
max_chol = df['chol'].max()
min_chol = df['chol'].min()
print("\nc) Cholesterol Levels:")
print("Maximum:", max_chol)
print("Minimum:", min_chol)
```

```python
# -----------------------------
# d) High Blood Pressure (>140)
# -----------------------------
high_bp = (df['trestbps'] > 140).sum()
print("\nd) Individuals with BP > 140:", high_bp)
```

```python
# -----------------------------
# e) High BP and Heart Disease
# -----------------------------
high_bp_hd = df[(df['trestbps'] > 140) & (df['target'] == 1)].shape[0]
print("\ne) High BP and Heart Disease:", high_bp_hd)
```

```python
# -----------------------------
# f) Histogram of Cholesterol
# -----------------------------
plt.figure()
plt.hist(df['chol'], bins=10)
plt.xlabel("Cholesterol")
plt.ylabel("Frequency")
plt.title("Cholesterol Distribution")
plt.show()

print("\nf) Most cholesterol values are clustered in the mid range.")
```

```python
# -----------------------------
# g) Bar chart: Chest Pain vs Heart Disease
# -----------------------------
cp_hd = df.groupby('cp')['target'].sum()

plt.figure()
cp_hd.plot(kind='bar')
plt.xlabel("Chest Pain Type")
plt.ylabel("Heart Disease Cases")
plt.title("Heart Disease by Chest Pain Type")
plt.show()
```

```python
# -----------------------------
# h) Lowest heart disease risk group
# -----------------------------
risk_group = df.groupby('cp')['target'].mean()
lowest_risk = risk_group.idxmin()
print("\nh) Lowest heart disease risk chest pain type:", lowest_risk)
```

```python
# -----------------------------
# i) Pie chart: Gender distribution
# -----------------------------
sex_counts = df['sex'].value_counts()

plt.figure()
sex_counts.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel("")
plt.title("Gender Distribution")
plt.show()
```

```python

# -----------------------------
# j) Percentage of males
# -----------------------------
male_percentage = (df['sex'] == 1).mean() * 100
print("\nj) Percentage of males:", round(male_percentage, 2), "%")
```

```python

# -----------------------------
# k) Scatter plot: Age vs Max Heart Rate
# -----------------------------
plt.figure()
plt.scatter(df['age'], df['thalach'])
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.title("Age vs Maximum Heart Rate")
plt.show()
```

```python

# -----------------------------
# l) Trend or clustering
# -----------------------------
print("\nl) Negative trend observed: heart rate decreases with increasing age.")
```

```python

# -----------------------------
# m) Box plot: Resting Blood Pressure
# -----------------------------
plt.figure()
plt.boxplot(df['trestbps'])
plt.ylabel("Resting Blood Pressure")
plt.title("Blood Pressure Distribution")
plt.show()
```

```python

# -----------------------------
# n) Median Blood Pressure
# -----------------------------
median_bp = df['trestbps'].median()
print("\nn) Median Blood Pressure:", median_bp)
```

```python

# -----------------------------
# o) Outliers
# -----------------------------
print("\no) Yes, outliers are visible in blood pressure values.")
```

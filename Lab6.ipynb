```python
#Task 01
from collections import deque
import heapq
graph1={
    0:[1,2],
    1:[3,4],
    2:[5,6],
    3:[],
    4:[],
    5:[],
    6:[]
}
#----------- BFS-------------
def bfs1(graph,start):
    visited=set()
    queue=deque([start])
    while queue:
        node=queue.popleft()
        if node not in visited:
            print(node,end=' ')
            visited.add(node)
            queue.extend(i for i in graph[node] if i not in visited)
# ------------DFS---------------
def dfs1(graph,start):
    visited=set()
    stack=[start]
    while stack:
        node=stack.pop()
        if node not in visited:
            print(node,end=' ')
            visited.add(node)
            stack.extend(reversed(graph[node]))
# ----------DLS-------------------
def dls1(graph,node,goal,limit,depth=0):
    print(f"Visiting:{node}, depth={depth}")
    if node==goal:
        print("Goal found")
        return True
    if depth==limit:
        return False
    for nbr in graph[node]:
        if dls1(graph,nbr,goal,limit,depth+1):
            return True
    return False
#-------------IDDFS-----------------
def iddfs1(graph,start,goal,max_depth):
    for depth in range(max_depth+1):
        print(f"\nDepth limit={depth}")
        if dls1(graph,start,goal,depth):
            print(f"Goal '{goal}' found")
            return True
    print("Goal not found")
    return False
#------------UCS----------------
def ucs1(graph,start,goal):
    pq=[(0,start)]
    visited=set()
    while pq:
        cost,node=heapq.heappop(pq)
        print(f"Visiting:{node}, cost={cost}")
        if node==goal:
            print(f"Goal reached with cost={cost}")
            return True
        if node not in visited:
            visited.add(node)
            for nbr in graph[node]:
                heapq.heappush(pq,(cost+1,nbr))
    return False

print("BFS: ",end='')
bfs1(graph1,1)
print("\nDFS :",end='')
dfs1(graph1,0)
print("\n\nDLS :")
dls1(graph1,0,6,2)
print("\nIDDFS :")
iddfs1(graph1,0,6,5)
print("\nUCS :")
ucs1(graph1,0,6)
```

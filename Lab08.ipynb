```python
# Task 01
import math
def alphabeta(depth,nodeIndex,isMax,scores,alpha,beta,targetDepth):
    if depth==targetDepth:
        return scores[nodeIndex]
    if isMax:
        best=-math.inf
        val1=alphabeta(depth+1,nodeIndex*2,False,scores,alpha,beta,targetDepth)
        best=max(best,val1)
        alpha=max(alpha,best)
        if beta<=alpha:
            return best  
        val2=alphabeta(depth+1,nodeIndex*2+1,False,scores,alpha,beta,targetDepth)
        best=max(best,val2)
        alpha=max(alpha,best)
        return best
    else:
        best=math.inf
        val1=alphabeta(depth+1,nodeIndex*2,True,scores,alpha,beta,targetDepth)
        best=min(best,val1)
        beta=min(beta,best)
        if beta<=alpha:
            return best  
        val2=alphabeta(depth+1,nodeIndex*2+1,True,scores,alpha,beta,targetDepth)
        best=min(best,val2)
        beta=min(beta,best)
        return best
scores=[4,2,-3,-6,7,0,5,8]
treeDepth=int(math.log2(len(scores)))
optimalValue=alphabeta(0,0,True,scores,-math.inf,math.inf,treeDepth)
print("The optimal value is:",optimalValue)
```

```python
# Task 02
import math
def evaluate(board):
    for i in range(3):
        if board[i][0]==board[i][1]==board[i][2]:
            return score(board[i][0])
        if board[0][i]==board[1][i]==board[2][i]:
            return score(board[0][i])
    if board[0][0]==board[1][1]==board[2][2]:
        return score(board[0][0])
    if board[0][2]==board[1][1]==board[2][0]:
        return score(board[0][2])
    return 0  
def score(player):
    if player=='X':
        return 10
    elif player=='O':
        return -10
    return 0
def isMovesLeft(board):
    for row in board:
        if ' ' in row:
            return True
    return False
def minimax(board,depth,isMax):
    scoreVal=evaluate(board)
    if scoreVal!=0 or not isMovesLeft(board):
        return scoreVal
    if isMax:
        best=-math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j]==' ':
                    board[i][j]='X'
                    best=max(best,minimax(board,depth+1,False))
                    board[i][j]=' '
        return best
    else:
        best=math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j]==' ':
                    board[i][j]='O'
                    best=min(best,minimax(board,depth+1,True))
                    board[i][j]=' '
        return best
def findWinner(board):
    val=evaluate(board)
    if val==10:
        return "X wins"
    elif val==-10:
        return "O wins"
    else:
        return "Draw"
board=[
    ['O','X','O'],
    ['O','X','X'],
    ['X','O','X']
]
optimalValue=minimax(board,0,True)
winner=findWinner(board)
print("The optimal value is:",optimalValue)
print("Result:",winner)
```
